```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

fig, ax = plt.subplots()
x_vals = np.linspace(-10, 10, 400)
y_vals = sigmoid(x_vals)
ax.plot(x_vals, y_vals, color='blue')
dot, = ax.plot([], [], 'ro')
text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

ax.set_xlim(-10, 10)
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel("Input")
ax.set_ylabel("Sigmoid Output")
ax.set_title("Sigmoid Activation Function - Animated Simulation")
ax.grid(True)

def init():
    dot.set_data([], [])
    text.set_text('')
    return dot, text

def animate(i):
    x = -10 + i * 0.2
    y = sigmoid(x)
    dot.set_data([x], [y])
    text.set_text(f'Input: {x:.2f}, Output: {y:.2f}')
    return dot, text

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=100, blit=True)
ani.save("sigmoid_activation_simulation.gif", writer='pillow')
```
